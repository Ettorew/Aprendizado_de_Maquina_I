```python
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np

df = pd.read_csv("arquivos_csv/treino.csv")

alvo = 'diag_hipertensao'
y_train = df[alvo]

X_train = df.drop(alvo, axis=1)

# Imprime o shape ANTES do balanceamento
print(f"Shape original de X_train: {X_train.shape}")
print(f"Contagem original de y_train: \n{y_train.value_counts()}")
print("-" * 30)

# 1. Inicializa o SMOTE
# random_state=42 garante que a criação de dados seja sempre a mesma
smote = SMOTE(random_state=42)

# 2. Aplica o SMOTE (fit_resample) APENAS nos dados de treino
# Ele cria novos X e y balanceados
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# 3. Verifica o resultado
print("\n--- DEPOIS DO SMOTE ---")
print(f"Shape de X_train_smote: {X_train_smote.shape}")
print(f"Contagem de y_train_smote: \n{y_train_smote.value_counts()}")

df_treino = X_train.copy()
df_treino[alvo] = y_train
df_treino.to_csv("arquivos_csv/treino_smote.csv", index=False)
```
